```python
%matplotlib inline
import matplotlib.pyplot as plt
import mpld3
import pandas as pd
import random
import seaborn as sns
mpld3.enable_notebook()
```

```python
data=pd.read_csv("Bollywood_actors.csv")
data.head()
```

```python
freq=1/12
print("min year ",min(data['Year']))
print("max year ",max(data['Year']))
x=data['Year']+freq*data['Month'] #x-axis corresponds to timeline
data["new_x"]=x #new column added to dataframe
actor=data['Actor_name'].unique()
color_list=[[0,0,0]] *len(actor)
for i in range(len(actor)): #list of colors to be used during scatter plot
    color_list[i]=[[random.uniform(0,1),random.uniform(0,1),random.uniform(0,1)]]
data.head()
```

```python
fig=plt.figure(figsize=(15,10))
plt.xlabel("Year")
plt.ylabel("Earning(in crores)")

ax=fig.add_subplot(111)
for i in range(len(actor)): #plot earnings of each actor in each iteration
    temp=data[data['Actor_name']==actor[i]]
    sc_temp=ax.scatter(temp['new_x'],temp['Gross_profit'],color=color_list[i],marker='o',label=actor[i])
    labels_temp=temp['Film_name'].values.tolist()
    labels= ['point {0}'.format(i + 1) for i in range(len(labels_temp))]
    tooltip = mpld3.plugins.PointLabelTooltip(sc_temp, labels=labels_temp)
    mpld3.plugins.connect(fig, tooltip)
plt.legend(loc="upper left")
mpld3.display()
```
